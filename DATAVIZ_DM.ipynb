```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import data from excel file  DATASET_ANNOTATION_position.xlsx
dataset_metadata = pd.read_excel('DATASET_ANNOTATION_position.xlsx', sheet_name='DATASET_ANNOTATION')

# Import measurements from excel file  gobbo_data.xlsx, sheet UD_Dat 
dataset_features = pd.read_excel('gobbo_data.xlsx', sheet_name='features')
```

```python
# Subset the metadata dataset to keep only the columns of interest
# Filter column kept by values 'TRUE' and  'OLIVER'
m1 = dataset_metadata.kept == 'TRUE'
m2 = dataset_metadata.kept == 'OLIVER'
m = m1 | m2
subset_metadata = dataset_metadata.loc[m]
```

## MODELOS DE BRINCADEIRA

```python
# Decision tree from sklearn
from sklearn.tree import DecisionTreeClassifier

# We want to predict the tag of the units based on the features
# We create a new dataset with the features and the tag
# We keep only the columns of interest

# Feature matrix
X = filtered.iloc[:, 15:41]

# Target vector
y = filtered.iloc[:, 8]

# Create label encoder
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

# Encode target vector
y = le.fit_transform(y)

# Print
y
```

```python
# Split dataset into training set and test set
# It must be a stratified split by tag

# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y, test_size=0.25) # 70% training and 30% test
```

```python
set(filtered.tag)
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
cross_val_score(clf, X, y, cv=10)
```

```python
# We apply a PCA to the feature matrix

# Import PCA
from sklearn.decomposition import PCA

# Create a PCA object
pca = PCA(n_components=X.shape[1])

# Fit the PCA and transform the data
X_pca = pca.fit_transform(X)

# Print
X_pca

# We plot the cumulative explained variance ratio

# Set figure size
plt.figure(figsize=(12, 5))

# Plot cumulative explained variance ratio
plt.plot(np.cumsum(pca.explained_variance_ratio_))

# Set title
plt.title('Cumulative explained variance ratio')
plt.show()

# We plot the first two principal components

# Set figure size
plt.figure(figsize=(12, 5))


# Plot first two principal components
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', s=50, alpha=0.5)

# Set title

plt.title('First two principal components')

plt.show()
```

```python
# We recreate the PCA object with 8 components

# Create a PCA object
pca = PCA(n_components=8)

# Fit the PCA and transform the data
X_pca = pca.fit_transform(X)
```

```python
# Now we run an LDA on the PCA-transformed data

# Import LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Create an LDA object
lda = LDA()

# Cross-validate the LDA

# Import cross_val_score
from sklearn.model_selection import cross_val_score

# Cross-validate the LDA with balanced accuracy
lda_cv = cross_val_score(lda, X_pca, y, cv=3, scoring='balanced_accuracy')

# Print
lda_cv

```

```python
# Random forest from sklearn
from sklearn.ensemble import RandomForestClassifier

# Create a random forest object
rf = RandomForestClassifier(n_estimators=100, random_state=0)

# Cross-validate rf with balanced accuracy
rf_cv = cross_val_score(rf, X, y, cv=10, scoring='balanced_accuracy')

# Print
rf_cv
```

```python
#https://towardsdatascience.com/visualizing-decision-trees-with-python-scikit-learn-graphviz-matplotlib-1c50b4aa68dc
```

```python
# Plot the decision tree for the Random Forest Classifier

# Import export_graphviz
from sklearn.tree import export_graphviz

# Import graphviz
import graphviz

rf.fit(X_train, y_train)

# Create a dot file
dot_data = export_graphviz(rf.estimators_[0], out_file=None, feature_names=X.columns, class_names=le.classes_, filled=True, rounded=True, special_characters=True)

# Create a graph from the dot file
graph = graphviz.Source(dot_data)

# Show the graph
graph
```

```python
# SVM from sklearn
from sklearn.svm import SVC

# Create an SVM object
svm = SVC(kernel='linear', C=1, random_state=0)

# Cross-validate svm with balanced accuracy
svm_cv = cross_val_score(svm, X, y, cv=10, scoring='balanced_accuracy')

# Print
svm_cv
```

```python
# Naïve Bayes from sklearn
from sklearn.naive_bayes import GaussianNB

# Create a Gaussian Naïve Bayes object
nb = GaussianNB()

# Cross-validate nb with balanced accuracy
nb_cv = cross_val_score(nb, X, y, cv=10, scoring='balanced_accuracy')

# Print
nb_cv
```

```python
# Plot decision tree for Decision Tree Classifier using tree from sklearn

# Import tree
from sklearn import tree
from sympy import n_order

# Create a decision tree object
dt = tree.DecisionTreeClassifier(max_depth=3, random_state=0)

# Fit the decision tree
dt.fit(X, y)


# Set figure size
plt.figure(figsize=(16, 8))

# Plot tree
tree.plot_tree(dt)
```
